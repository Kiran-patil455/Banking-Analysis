```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector
# Connect to server
cnx = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="root")
```

```python
query = " SELECT * FROM banking_case.customer"
```

```python
df = pd.read_sql(query, cnx)
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

#Univariate Analysis
plt.figure(figsize=(15,10))

for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.tight_layout()    
plt.show()
```

## Heatmap

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
correlation_matrix = df[numerical_cols].corr()

plt.figure.figzize=(12,12)
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

# Insights of  EDA
1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
2. Among all continents, European customers hold the highest average number of credit cards. This suggests a more credit-oriented financial behavior in Europe, potentially driven by widespread access to banking services, higher levels of financial literacy, and a stronger reliance on credit-based purchasing compared to other regions.
3. The majority of customers own only one credit card, significantly outnumbering those with two or three credit cards. This indicates a preference for simpler credit management and possibly a cautious approach toward credit usage among the customer base.
4. Private banking customers hold the highest number of credit cards on average compared to other customer segments. This trend reflects their higher financial engagement and access to a broader range of financial products, often tailored to meet the needs of high-net-worth individuals.
